```python
import matplotlib.pyplot as plt
import numpy as np

# 创建图形和子图
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 15))
fig.suptitle('Heat Transfer Diagram', fontsize=16)

# 顶部图：建筑剖面图
def draw_building(ax):
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 5)
    ax.axis('off')
    
    # 画建筑轮廓
    ax.add_patch(plt.Rectangle((2, 0), 6, 4, fill=False))
    ax.add_patch(plt.Rectangle((3, 0), 1, 4, fill=False))
    
    # 添加文字和箭头
    ax.text(0.5, 3, 'air', fontsize=12)
    ax.text(9, 4, 'Tair', fontsize=12)
    for i in range(3):
        ax.arrow(0.5, 2.5-i*0.5, 1, 0, head_width=0.1, head_length=0.1)
    
    # 添加温度点
    for i, label in enumerate(['T1', 'T2', 'T3']):
        ax.text(4+i*2, 2, label, fontsize=10)
    
    ax.text(2.5, -0.5, 'plate', fontsize=10)
    ax.text(5, -0.5, 'main heat', fontsize=10)

# 中间图：电路图
def draw_circuit(ax):
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 5)
    ax.axis('off')
    
    # 画主要电路
    ax.plot([1, 9], [2.5, 2.5], 'k-')
    ax.plot([3, 3], [2.5, 1.5], 'k-')
    ax.plot([5, 5], [2.5, 1.5], 'k-')
    ax.plot([7, 7], [2.5, 1.5], 'k-')
    
    # 添加电阻符号
    for x in [2, 4, 6, 8]:
        ax.plot([x, x+0.5], [2.5, 3], 'k-')
        ax.plot([x+0.5, x+1], [3, 2.5], 'k-')
    
    # 添加标签
    ax.text(0.5, 2.5, 'T1', fontsize=10)
    ax.text(9.5, 2.5, 'Tair', fontsize=10)
    ax.text(3, 1, 'L/k·A', fontsize=10)
    ax.text(5, 1, 'L/k·A', fontsize=10)
    ax.text(7, 1, '1/h·A', fontsize=10)

# 底部图：简化电路
def draw_simplified_circuit(ax):
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 2)
    ax.axis('off')
    
    # 画主线
    ax.plot([1, 9], [1, 1], 'k-')
    
    # 添加电阻符号
    for x in [2.5, 4.5, 6.5]:
        ax.plot([x, x+0.5], [1, 1.5], 'k-')
        ax.plot([x+0.5, x+1], [1.5, 1], 'k-')
    
    # 添加标签
    labels = ['T1', 'T2', 'T3', 'Tair']
    for i, label in enumerate(labels):
        ax.text(1.5 + i*2, 0.5, label, fontsize=10)

# 绘制三个子图
draw_building(ax1)
draw_circuit(ax2)
draw_simplified_circuit(ax3)

plt.tight_layout()
plt.show()
```
